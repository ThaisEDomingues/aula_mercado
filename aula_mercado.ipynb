```python
# O nosso desafio é mostrar os produtos de um mercado
# Escolher os produtos para ver info
# Adicionar novos produtos no mercado
# Excluir um produto do mercado
#-----------------------------------------------------

# Criar as variáveis globais para o software
lista_produtos = ['Detergente Ypê',
                  'Refrigerante Gengibirra',
                  'Arroz',
                  'Feijão',
                  'Água mineral']

lista_preco = [2.50, 5, 7.25, 4, 3.45]
```

```python
# Para criar uma função devemos armazenar os códigos dentro de um bloco e nomear
# def comando para definir uma função ex: def nomefuncao ():
# No comando print (f'') - não pode ter espaço entre o f e as aspas
# No hora de você atribuir o valor input você informa o tipo da variável antes do comando input
def  ver_produtos():
  print(f'Os produtos são: {lista_produtos}')
  print(f'e seus preços são: {lista_preco}')

def adicionar_produtos():
  novo_produto = input('Informe  nome do produto: ')
  novo_preco = float(input('Informe o preço: '))

  # /\ CHAMAMOS A LISTA - ATRIBUIMOS A FUNÇÃO APPEND (ADICIONA) -
  # COLOCAMOS O QUE VAI SER ADICIONADO
  lista_produtos.append(novo_produto)
  lista_preco.append(novo_preco)
```

```python
# Executar a função que mostra as listas de produtos e de preços
# Executar a função que adiciona um produtos
# Executar a função que exclui um produto

print('----MENU DO MERCADO----')
print('Escolha a opção: ')
print('1 - Ver produtos: ')
print('2 - Cadastrar novo produto: ')
#print('3 - Excluir produto: ')

opcao = int(input('Informe a opção desejada: '))

if opcao == 1:
   ver_produtos()
#else + if ou seja elif
elif opcao == 2:
     adicionar_produtos()
     ver_produtos()

else:
  print('Opção não encontrada. ')
```
